```python
train['Discount'] = train['Discount'].map({"Yes":1, "No":0})
train.head()
```

```python
train[['Store_id', 'Holiday', 'Discount']] = train[['Store_id', 'Holiday', 'Discount']].astype(object)
train['Date'] = pd.to_datetime(train.Date)
```

```python
train[['Store_id','Holiday', 'Discount']] = train[['Store_id', 'Holiday', 'Discount']].astype(int)
```

```python
from sklearn.model_selection import train_test_split
X = train.drop(['Sales'],axis=1)
y = train['Sales']
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import LabelEncoder
e1 = LabelEncoder()
X_train['Store_Type'] = e1.fit_transform(X_train['Store_Type'])
X_valid['Store_Type'] = e1.transform(X_valid['Store_Type'])

e2 = LabelEncoder()
X_train['Location_Type'] = e2.fit_transform(X_train['Location_Type'])
X_valid['Location_Type'] = e2.transform(X_valid['Location_Type'])

e3 = LabelEncoder()
X_train['Region_Code'] = e3.fit_transform(X_train['Region_Code'])
X_valid['Region_Code'] = e3.transform(X_valid['Region_Code'])
```

```python
X_train.head()
```

```python
X_train.corr()
```

```python
X_train = X_train.drop('Date', axis=1)
X_valid = X_valid.drop('Date', axis=1)
```

```python
X_train.info()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression, Ridge,Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
def eval_model(true, predicted):
  mse = mean_squared_error(true, predicted)
  mae = mean_absolute_error(true, predicted)
  r2 = r2_score(true, predicted)
  rmse = np.sqrt(mse)
  return mae, r2, rmse
```

```python
models = {
    'Linear Regression': LinearRegression(),
    'Ridge': Ridge(),
    'Lasso': Lasso(),
    'Random Forest': RandomForestRegressor(),
    'K Neighbours Regressor': KNeighborsRegressor(),
    'Decision Tree': DecisionTreeRegressor()
}

for i in range(len(models)):
  model = list(models.values())[i]
  model.fit(X_train, y_train)

  y_train_pred = model.predict(X_train)
  y_test_pred = model.predict(X_valid)

  model_train_mae, model_train_r2, model_train_rmse = eval_model(y_train, y_train_pred)
  model_test_mae, model_test_r2, model_test_rmse = eval_model(y_valid, y_test_pred)

  print('\n\n')
  print(list(models.keys())[i])
  print('Model performance for training set:\n')
  print(f'Mean Absolute Error: {model_train_mae}')
  print(f'Root Mean Absolute Error: {model_train_rmse}')
  print(f'R2 score: {model_train_r2}')
  print('-----------------------------------------------')
  print('Model performance for testing set:\n')
  print(f'Mean Absolute Error: {model_test_mae}')
  print(f'Root Mean Absolute Error: {model_test_rmse}')
  print(f'R2 score: {model_test_r2}')
  print('-----------------------------------------------')
```
